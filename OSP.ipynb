```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
# if using a Jupyter notebook, includue:
%matplotlib inline
```

```python
sheet1 = pd.read_csv(r'../submission/Sched/FCFS.txt',sep= "\s+")
sheet2 = pd.read_csv(r'../submission/Sched/SJF.txt',sep= "\s+")
sheet3 = pd.read_csv(r'../submission/Sched/RR.txt',sep= "\s+")
```

```python
bins_list = [0, 200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800]
df11 = pd.DataFrame(sheet1, columns= ['W']), "r", "FCFS Waiting Time"
df12 = pd.DataFrame(sheet1, columns= ['T']), "r", "FCFS Turn Around Time"
df21 = pd.DataFrame(sheet2, columns= ['W']), "g", "SJF Waiting Time"
df22 = pd.DataFrame(sheet2, columns= ['T']), "g", "SJF Turn Around Time"
df31 = pd.DataFrame(sheet3, columns= ['W']), "b", "RR Waiting Time"
df32 = pd.DataFrame(sheet3, columns= ['T']), "b", "RR Turn Around Time"
dfs1 = df11, df21, df31
dfs2 = df12, df22, df32
```

```python
plt.figure(figsize=(8,8))
for df in dfs1:
    ax = sns.distplot(df[0], color=df[1], hist=False, kde_kws={'shade':True},
                      label=df[2], axlabel="Waiting Time Comparison", bins=bins_list)
    ax.set(xlim=(0, 1800))
    ax.set(ylim=(0, 0.005))
```

```python
for df in dfs1:
    print(df[0].describe())
```

```python
plt.figure(figsize=(8,8))
for df in dfs2:
    ax = sns.distplot(df[0], color=df[1], hist=False, kde_kws={'shade':True},
                      label=df[2], axlabel="Turn Around Time Comparison", bins=bins_list)
    ax.set(xlim=(0, 1800))
    ax.set(ylim=(0, 0.005))
```

```python
for df in dfs2:
    print(df[0].describe())
```

```python
plt.figure(figsize=(8,8))
ax = sns.distplot(df11[0], kde=False, norm_hist=False,
                  bins=bins_list, color=df11[1], axlabel=df11[2],
                  hist_kws = {'edgecolor':'k', 'linewidth': 1, 'align': 'mid'})
ax.set(xlim=(0, 1800))
ax.set(ylim=(0, 160))
for rect in ax.patches:
    height = rect.get_height()
    if height != 0:
        ax.annotate(f'{int(height)}', xy=(rect.get_x()+rect.get_width()/2, height),
                    xytext=(0, 5), textcoords='offset points', ha='center', va='bottom')
```
